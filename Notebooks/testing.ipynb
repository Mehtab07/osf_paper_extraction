```python
import fitz  # PyMuPDF

doc = fitz.open("test/2sz48.pdf")
full_text = ""
for page in doc:
    full_text += page.get_text()
doc.close()
```

```python
import re

# Define common section headers
section_headers = ["Abstract", "Introduction", "Related Work", "Method", "Methods", "Results", "Discussion", "Conclusion", "References"]

# Build a regex pattern
pattern = "|".join([rf"\n\s*{header}\s*\n" for header in section_headers])
sections = re.split(pattern, full_text, flags=re.IGNORECASE)

# Optional: match headers for naming
headers_found = re.findall(pattern, full_text, flags=re.IGNORECASE)
```

```python
with open("output_sections.md", "w", encoding="utf-8") as f:
    for header, content in zip(headers_found, sections[1:]):  # [1:] skips anything before first header
        f.write(f"## {header.strip()}\n\n{content.strip()}\n\n")
```
