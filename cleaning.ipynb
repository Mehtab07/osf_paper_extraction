```python
project_df = pd.read_csv('project_ids.csv')
my_ids_df = pd.read_csv('my_id.csv')

my_ids = set(my_ids_df['id'].dropna().astype(str))
project_ids = set(project_df['ids'].dropna().astype(str))

matched_df = project_df[project_df['ids'].isin(my_ids)]
matched_df.to_csv('doi.csv', index=False)

found_ids = list(my_ids.intersection(project_ids))
not_found_ids = list(my_ids.difference(project_ids))

status_df = pd.DataFrame({
    'Found': pd.Series(found_ids),
    'Not Found': pd.Series(not_found_ids)
})
status_df.to_csv('match_status.csv', index=False)
```

```python
import pandas as pd
import requests
import time

def get_osf_title(project_id):
    url = f"https://api.osf.io/v2/nodes/{project_id}/"
    try:
        response = requests.get(url)
        response.raise_for_status()
        return response.json().get("data", {}).get("attributes", {}).get("title", None)
    except Exception as e:
        print(f"Error fetching title for {project_id}: {e}")
        return None

def add_titles_to_csv(input_csv_path, output_csv_path):
    df = pd.read_csv(input_csv_path)

    if "id" not in df.columns:
        raise ValueError("CSV must contain a column named 'id'.")

    df["osf_title"] = df["id"].apply(lambda pid: get_osf_title(str(pid)))
    df.to_csv(output_csv_path, index=False)
    print(f"✅ Saved updated CSV with titles to: {output_csv_path}")

if __name__ == "__main__":
    input_csv = "my_id.csv"                      
    output_csv = "my_id_with_titles.csv"         
    add_titles_to_csv(input_csv, output_csv)
```

```python
import pandas as pd
import os

input_csv = "semantic_search_results.csv"
pdf_folder = "papers"
downloaded_csv = "downloaded.csv"
remaining_csv = "remaining.csv"

# Read the input CSV
df = pd.read_csv(input_csv)

# Get all PDF filenames (without .pdf)
downloaded_ids = {os.path.splitext(f)[0] for f in os.listdir(pdf_folder) if f.endswith(".pdf")}

# Filter rows
downloaded_df = df[df["id"].astype(str).isin(downloaded_ids)]
remaining_df = df[~df["id"].astype(str).isin(downloaded_ids)]

# Save results
downloaded_df.to_csv(downloaded_csv, index=False)
remaining_df.to_csv(remaining_csv, index=False)

print(f" {len(downloaded_df)} entries saved to '{downloaded_csv}'")
print(f"{len(remaining_df)} entries saved to '{remaining_csv}'")
```
